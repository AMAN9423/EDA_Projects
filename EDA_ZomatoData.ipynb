```python
df_country = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Input Data/Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df_final = pd.merge(df, df_country, on='Country Code', how='left',)
df_final.head()
```

**Check Datatype**

```python
country_names = df_final.Country.value_counts().index
country_names
```

```python
country_values = df_final.Country.value_counts().values
```

```python
##Pie Chart
plt.pie(country_values[:3], labels=country_names[:3], autopct="%1.2f%%")
plt.show()
```

**Observation** : Zomato maximum records or transaction are from India.
After that USA and the UK

```python
df_final.columns
```

```python
ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

**Observations** :

1.   When rating is between 4.5 to 4.9----> Excellent
2.   When rating is between 4.0 to 4.4----> Very Good

1.   When rating is between 3.5 to 3.9----> Good
2.   When rating is between 2.5 to 2.9----> Average

1.   When rating is between 2.0 to 2.4----> Poor






```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,6)
sns.barplot(x='Aggregate rating', y = 'Rating Count', data=ratings, hue='Rating color', palette=['white','red','orange','yellow','green','teal'])
plt.show()
```

**Observations** :

1.   Not rated count is very high
2.   maximum number of ratings are between 2.5 to 3.4



```python
##Count Plot
sns.countplot(x='Rating color', data=ratings, palette=['white','red','orange','yellow','green','teal'])
```
